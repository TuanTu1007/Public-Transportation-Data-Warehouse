```python
import re


with open('df_cleaned.csv', 'r', encoding='utf-8') as f:
    data = f.read()

# Sử dụng regex để tìm các chuỗi trong dấu ngoặc kép và loại bỏ dấu phẩy bên trong
modified_data = re.sub(r'\"(.*?)\"', lambda match: match.group(0).replace(',', ''), data)

with open('df_cleaned_final.csv', 'w', encoding='utf-8') as f:
    f.write(modified_data)
```

```python
import csv
with open('df_cleaned_final.csv', 'r', encoding='utf-8') as infile, open('df_cleaned_final1.csv', 'w', newline='', encoding='utf-8') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # Duyệt qua từng hàng
    for row in reader:
        # Loại bỏ dấu nháy kép trong từng cột của hàng
        new_row = [col.replace('"', '') for col in row]
        # Ghi hàng đã xử lý vào file mới
        writer.writerow(new_row)
```
